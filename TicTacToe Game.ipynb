```python
def printer(val):
    print("    |     |    \n {0}  | {1}   | {2}   \n    |     |    \n---------------\n    |     |    \n {3}  | {4}   | {5}   \n    |     |    \n---------------\n    |     |    \n {6}  | {7}   | {8}   \n    |     |    \n---------------".format(str(val[0]),val[1],val[2],val[3],val[4],val[5],val[6],val[7],val[8]))
```

```python
def plotter(player_list,input1,char):
    if(len(player_list)==0):
        player_list.append([' ',' ',' ',' ',' ',' ',' ',' ',' '])
    if(player_list[(input1-1)] == ' '):
        player_list.insert(input1-1,char)
    return player_list
    
```

```python
def win_func(list_val):
    if(list_val == [' ',' ',' ',' ',' ',' ',' ',' ',' ']):
        return False
    
    val = 0
    ## horizontal match
    for i in range(3):
        if(list_val[0+val] == list_val[1+val] == list_val[2+val] != ' '):
            return True
        val = 3
    val = 0    
    ## vertical match
    for i in range(3):
        if(list_val[0+val] == list_val[3+val] == list_val[6+val]  != ' '):
            return True
        val = 1
     
    ## diagonal match match
    if(list_val[0] == list_val[4] == list_val[8]  != ' '):
            return True
    if(list_val[2] == list_val[4] == list_val[6]  != ' '):
            return True
    return False
        
    
```

```python
player = list([' ',' ',' ',' ',' ',' ',' ',' ',' '])
pl1 = 'X'
pl2 = 'O'
chance = 1
dec = 'first_time'
while True:
    printer(player)
    if(chance == 1):
        a = int(input('Player-1: Enter you choice'))
        player = plotter(player,a,pl1)
        printer(player)
        chance = 2
        if(win_func(player)):
            print('Player 1 WON !!' )
            dec = input("Want to continue Yes/No")
            
    if(chance == 2):
        a2 = int(input('Player-2: Enter you choice'))
        player = plotter(player,a2,pl2)
        chance = 1
        if(win_func(player)):
            print('Player 2 WON !!' )
            dec = input("Want to continue Yes/No")
        
    if(dec == 'Yes'):
        printer([' ',' ',' ',' ',' ',' ',' ',' ',' '])
        player1 = list([' ',' ',' ',' ',' ',' ',' ',' ',' '])
        player2 = list([' ',' ',' ',' ',' ',' ',' ',' ',' '])
        pl1 = 'X'
        pl2 = 'O'
        chance = 1
        dec = 'No'
    if(dec == 'No'):
        print("Thanks for playing ankits TIc Tac Toe")
        break
        
        
        
    
    
    
```
